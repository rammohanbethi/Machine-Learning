# Multiple plots on tha same canvas

```python
plt.subplot(2,2,1) #plot(num_row, num_col,plot_number)
plt.plot(x,y,'r',marker='*')
plt.subplot(2,2,2)
plt.plot(y,x,'g--')
plt.subplot(2,2,3)
plt.plot(x,y,'b+')
plt.subplot(2,2,4)
plt.plot(y,x,'y*-')
plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width,height
axes2 = fig.add_axes([0.2,0.5,0.4,0.4])

axes1.plot(x,y,'r')
axes1.set_xlabel("x-main")
axes1.set_ylabel("y-main")
axes1.set_title("Main Graph")

axes2.plot(y,x,'g')
axes2.set_xlabel("x-small")
axes2.set_ylabel("y-small")
axes2.set_title("small Graph")
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label='x**2',color='m')
ax.plot(x,x**3,label='x**3',color='b')
ax.legend()
```
